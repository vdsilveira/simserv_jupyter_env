## Objetivo

Verificar se existe uma IA local acessível pelo notebook e listar modelos disponíveis.

No SimServ, a IA local pode ser fornecida por um servidor Ollama. Dependendo da configuração do momento, ela pode estar disponível pelo endereço `http://ollama:11434`.

```python
# Teste: verifica se há LLM disponível (Ollama Cloud ou local) e lista modelos.

import json
import os
from urllib.request import urlopen, Request
from urllib.error import URLError, HTTPError

base_url_ollama = os.getenv('OLLAMA_BASE_URL', 'https://api.ollama.com').rstrip('/')
api_key = os.getenv('OLLAMA_API_KEY')

headers = {'User-Agent': 'Mozilla/5.0', 'Accept': 'application/json'}
if api_key:
    headers['Authorization'] = f'Bearer {api_key}'

urls = [base_url_ollama]
for fb in ['http://ollama:11434', 'http://127.0.0.1:11434', 'http://localhost:11434']:
    if fb not in urls:
        urls.append(fb)

resposta_json = None
url_funcionando = None

for url_base in urls:
    url = f"{url_base}/api/tags"
    try:
        req = Request(url, headers=headers)
        with urlopen(req, timeout=5) as resposta:
            texto = resposta.read().decode("utf-8")
            dados = json.loads(texto)
            if 'models' in dados:
                resposta_json = dados
                url_funcionando = url
                break
    except Exception as erro:
        print(f"Não respondeu em {url}: {erro}")

if resposta_json is None:
    print("\nNão foi possível acessar a IA (Ollama Cloud ou local) neste momento.")
    print("Isso não impede o uso normal do JupyterLab e dos notebooks Python.")
else:
    print("\nServidor de IA respondeu corretamente em:", url_funcionando)
    modelos = [m.get("name", "sem_nome") for m in resposta_json.get("models", [])]
    if modelos:
        print(f"Modelos disponíveis ({len(modelos)}):")
        for modelo in modelos:
            print(" -", modelo)
    else:
        print("O servidor respondeu, mas nenhum modelo foi listado.")
```

## Uso didático da IA

A IA pode ajudar a:

- explicar um erro;
- sugerir uma correção mínima;
- comentar um trecho de código;
- transformar uma narrativa simples em pseudocódigo;
- sugerir perguntas para explorar um resultado.

Mas a IA não substitui a análise do aluno. Sempre compare a resposta da IA com o que o código realmente fez.
