
## 8. Consulta opcional ร  IA local do SimServ

A prรณxima cรฉlula tenta encontrar um servidor Ollama disponรญvel. Em algumas instalaรงรตes do SimServ, a IA local pode estar acessรญvel como:

- `http://ollama:11434`
- `http://localhost:11434`
- `http://127.0.0.1:11434`

Se nรฃo funcionar, copie o prompt salvo em arquivo e use outro meio indicado pelo professor ou monitor.


````python
# 9. Funรงรตes para consultar o Ollama, se estiver disponรญvel

import requests

URLS_OLLAMA = [
    "http://ollama:11434",
    "http://localhost:11434",
    "http://127.0.0.1:11434",
]


def encontrar_ollama(urls=URLS_OLLAMA, timeout=5):
    for base_url in urls:
        try:
            resp = requests.get(f"{base_url}/api/tags", timeout=timeout)
            if resp.ok:
                dados = resp.json()
                modelos = [m.get("name") for m in dados.get("models", []) if m.get("name")]
                return base_url, modelos
        except Exception:
            pass
    return None, []


def escolher_modelo(modelos):
    if not modelos:
        return None
    preferencias = ["qwen", "llama", "gemma", "deepseek", "mistral"]
    for pref in preferencias:
        for modelo in modelos:
            if pref in modelo.lower():
                return modelo
    return modelos[0]


def consultar_ollama(prompt, modelo=None, base_url=None, timeout=180):
    if base_url is None or modelo is None:
        raise RuntimeError("base_url e modelo precisam ser definidos.")
    payload = {
        "model": modelo,
        "prompt": prompt,
        "stream": False,
        "options": {
            "temperature": 0.2
        }
    }
    resp = requests.post(f"{base_url}/api/generate", json=payload, timeout=timeout)
    resp.raise_for_status()
    return resp.json().get("response", "")

base_url_ollama, modelos_disponiveis = encontrar_ollama()
modelo_escolhido = escolher_modelo(modelos_disponiveis)

print("Ollama encontrado em:", base_url_ollama)
print("Modelos disponรญveis:", modelos_disponiveis)
print("Modelo escolhido:", modelo_escolhido)
````

````python
# 10. Rodar a consulta ao LLM
# Se esta cรฉlula falhar por falta de modelo ou servidor, use o arquivo de prompt salvo na pasta de saรญda.

resposta_llm = None

if base_url_ollama is None or modelo_escolhido is None:
    print("IA local nรฃo encontrada ou nenhum modelo disponรญvel.")
    print("Use o arquivo de prompt salvo em:")
    print(arquivo_prompt)
else:
    print(f"Consultando modelo {modelo_escolhido} em {base_url_ollama}...")
    resposta_llm = consultar_ollama(
        prompt_exequibilidade,
        modelo=modelo_escolhido,
        base_url=base_url_ollama,
        timeout=240
    )
    print("Resposta recebida. Primeira prรฉvia:\n")
    print(resposta_llm[:4000])
    if len(resposta_llm) > 4000:
        print("\n[Resposta truncada na tela. A cรฉlula seguinte salva o relatรณrio completo.]")
````
