```python
import importlib.util
import importlib.metadata
import sys
import platform
from pathlib import Path
import getpass
import datetime

def verificar_pacote(modulo, pacote=None):
    pacote = pacote or modulo
    encontrado = importlib.util.find_spec(modulo) is not None

    versao = "não disponível"
    if encontrado:
        try:
            versao = importlib.metadata.version(pacote)
        except importlib.metadata.PackageNotFoundError:
            versao = "módulo encontrado, versão não identificada"

    return {
        "modulo": modulo,
        "pacote": pacote,
        "encontrado": encontrado,
        "versao": versao
    }


def health_check_jupyter():
    pacotes = {
        "jupyterlab_lsp": "jupyterlab-lsp",
        "jupyter_lsp": "jupyter-lsp",
        "pylsp": "python-lsp-server",
        "jedi": "jedi",
        "numpy": "numpy",
        "pandas": "pandas",
        "matplotlib": "matplotlib",
        "IPython": "ipython",
        "ipykernel": "ipykernel",
        "requests": "requests",
    }

    linhas = []

    linhas.append("HEALTH CHECK - AMBIENTE PYTHON/JUPYTER")
    linhas.append("=" * 42)
    linhas.append(f"Data/hora: {datetime.datetime.now()}")
    linhas.append(f"Usuário: {getpass.getuser()}")
    linhas.append(f"Python: {sys.version}")
    linhas.append(f"Executável Python: {sys.executable}")
    linhas.append(f"Sistema: {platform.platform()}")
    linhas.append(f"Pasta atual: {Path.cwd()}")
    linhas.append(f"HOME: {Path.home()}")
    linhas.append("")

    linhas.append("PACOTES VERIFICADOS")
    linhas.append("-" * 25)

    for modulo, pacote in pacotes.items():
        resultado = verificar_pacote(modulo, pacote)

        status = "OK" if resultado["encontrado"] else "NÃO ENCONTRADO"

        linhas.append(
            f"{resultado['pacote']:25s} | "
            f"módulo: {resultado['modulo']:18s} | "
            f"{status:15s} | "
            f"versão: {resultado['versao']}"
        )

    linhas.append("")
    linhas.append("INTERPRETAÇÃO")
    linhas.append("-" * 15)
    linhas.append("OK significa que o módulo foi encontrado no ambiente Python atual.")
    linhas.append("NÃO ENCONTRADO indica que o pacote provavelmente precisa ser instalado.")
    linhas.append("A versão pode não aparecer quando o nome do módulo é diferente do nome do pacote.")

    return "\n".join(linhas)


relatorio = health_check_jupyter()
print(relatorio)
```
