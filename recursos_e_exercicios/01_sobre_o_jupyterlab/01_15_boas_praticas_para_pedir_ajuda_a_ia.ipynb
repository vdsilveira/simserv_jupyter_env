## Objetivo

Aprender a escrever pedidos claros para a IA quando estiver estudando programação, gráficos ou simulações.

## Pedido fraco

```text
Meu código não funciona. Corrija.
```

Problema: falta contexto. A IA não sabe o objetivo, o erro, a parte importante do código nem o nível de explicação desejado.

## Pedido melhor

```text
Estou aprendendo Python no JupyterLab.
Explique o erro abaixo em linguagem simples.
Depois sugira uma correção mínima.
Não reescreva o programa inteiro.

Erro:
[cole aqui a mensagem de erro]

Código:
[cole aqui o trecho de código]
```

````python
erro_exemplo = "NameError: name 'x' is not defined"
codigo_exemplo = "print(x)"

prompt_recomendado = f"""
Estou aprendendo Python no JupyterLab.
Explique o erro abaixo em linguagem simples.
Depois sugira uma correção mínima.
Não reescreva o programa inteiro.

Erro:
{erro_exemplo}

Código:
{codigo_exemplo}
""".strip()

print(prompt_recomendado)
````

## Estrutura útil para pedir ajuda

| Informação | Exemplo |
|---|---|
| Meu objetivo | Quero desenhar um gráfico de tempo versus posição. |
| Meu nível | Estou começando em Python. |
| O erro | Copiar a mensagem completa. |
| O trecho relevante | Colar apenas a parte necessária do código. |
| O tipo de resposta desejado | Correção mínima e explicação passo a passo. |

## Regra de ouro

Peça para a IA explicar o raciocínio **em termos didáticos**, mas confira executando o código e observando os resultados.
