## Identificação dos comandos principais

| Referência | Ícone ou texto aproximado | Função |
|---|---|---|
| A | `+` | Criar nova célula abaixo da célula selecionada. |
| B | tesoura | Recortar célula selecionada. |
| C | dois retângulos | Copiar célula selecionada. |
| D | prancheta | Colar célula copiada ou recortada. |
| E | triângulo `▶` | Executar a célula selecionada. |
| F | quadrado `■` | Interromper execução. Útil quando uma célula demora demais. |
| G | seta circular | Reiniciar o kernel. Use quando o notebook travar ou ficar confuso. |
| H | setas duplas | Executar células em sequência, dependendo da opção escolhida. |
| I | menu `Code` | Escolher tipo da célula: código, Markdown etc. |
| J | `Open in...` | Abrir o arquivo em outro modo ou visualizador, quando disponível. |
| K | `notebook is read-only` | Aviso de que o notebook está em pasta de somente leitura. |
| L | `Python 3 (ipykernel)` | Kernel ativo que executa o código Python. |
