## Objetivo

Usar variáveis para representar grandezas e fazer cálculos simples.

```python
massa = 2.0
aceleracao = 9.8
forca = massa * aceleracao

print("Massa:", massa, "kg")
print("Aceleração:", aceleracao, "m/s²")
print("Força:", forca, "N")
```

```python
temperatura_celsius = 25
temperatura_kelvin = temperatura_celsius + 273.15

print("Temperatura em Celsius:", temperatura_celsius)
print("Temperatura em Kelvin:", temperatura_kelvin)
```

## Texto formatado

O `f` antes das aspas permite inserir valores de variáveis dentro do texto.

```python
nome = "Aluno"
atividade = "primeiro gráfico"

print(f"{nome} está preparando o {atividade} no SimServ.")
```
