# 01_04 — Como copiar notebook para a área de trabalho

> **Material de familiarização com o JupyterLab do SimServ.**  
> Estes notebooks foram pensados para ficar em uma pasta de somente leitura, por exemplo `/sobre_ambientes_simserv`.  
> Para editar, testar livremente ou salvar resultados, faça uma cópia para sua área de trabalho, como `/home/jovyan/work` ou `/home/jovyan/laboratorio`.

## Objetivo

Aprender o fluxo correto para usar notebooks fornecidos pelo professor.

## Procedimento recomendado

1. Abra o notebook original na pasta de orientação ou de materiais.
2. Veja se aparece a mensagem **`notebook is read-only`**.
3. Use o menu **File > Save Notebook As...**.
4. Salve a cópia em `/home/jovyan/work` ou `/home/jovyan/laboratorio`.
5. Use um nome simples, sem espaços e sem acentos.

Exemplo de nome:

```text
experimento1_nome_do_aluno_V1.ipynb
```

````python
from pathlib import Path
import os

def encontrar_area_de_trabalho():
    """Procura uma pasta gravável para salvar resultados do aluno."""
    candidatas = [
        Path.home() / "ambiente_laboratorio",
        Path.home() / "laboratorio",
        Path.home() / "work",
        Path("/home/jovyan/laboratorio"),
        Path("/home/jovyan/work"),
        Path.cwd(),
    ]
    for pasta in candidatas:
        try:
            if pasta.exists() and os.access(pasta, os.W_OK):
                return pasta
        except Exception:
            pass
    raise RuntimeError(
        "Não encontrei uma pasta gravável. Abra uma cópia do notebook na sua área de trabalho."
    )

AREA_TRABALHO = encontrar_area_de_trabalho()
print("Área de trabalho escolhida:", AREA_TRABALHO)
````

## Teste seguro de escrita

A célula abaixo cria um pequeno arquivo de teste na área de trabalho. Ela não altera o notebook original.

````python
from datetime import datetime

arquivo_teste = AREA_TRABALHO / "teste_escrita_simserv.txt"
arquivo_teste.write_text(
    f"Teste de escrita no SimServ realizado em {datetime.now()}\n",
    encoding="utf-8"
)

print("Arquivo criado com sucesso:")
print(arquivo_teste)
````
