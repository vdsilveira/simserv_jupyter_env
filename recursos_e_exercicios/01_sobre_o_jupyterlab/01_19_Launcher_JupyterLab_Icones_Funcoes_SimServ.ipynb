
# Guia do Launcher do JupyterLab no SimServ

**Tema:** o que fazem os ícones do *Launcher* do JupyterLab.

Este notebook foi preparado para alunos que estão começando a usar o JupyterLab no SimServ.  
Ele explica, em linguagem didática, os principais ícones que aparecem no Launcher:

- **Notebook**: `Python 3 (ipykernel)` e, quando disponível, `Desktop`;
- **Console**: `Python 3`;
- **Other**: `Terminal`, `Text File`, `Markdown File`, `Python File`, `Chat` e `Show Contextual Help`.

> **Ideia central:** o Launcher não é um “programa único”; ele é uma tela de atalhos para criar ou abrir diferentes tipos de ambientes de trabalho dentro do JupyterLab.



## 1. Como abrir o Launcher

Normalmente há três formas simples:

1. Clique no botão **`+`** no painel lateral esquerdo.
2. Use o menu **File → New Launcher**.
3. Feche ou minimize abas abertas até encontrar a aba chamada **Launcher**.

O Launcher costuma ser dividido em seções, por exemplo:

- **Notebook**
- **Console**
- **Other**

Os nomes exatos podem mudar conforme o ambiente instalado no servidor.



## 2. Visão geral dos ícones

| Seção do Launcher | Ícone / opção | O que abre | Para que serve |
|---|---|---|---|
| Notebook | **Python 3 (ipykernel)** | Um arquivo `.ipynb` | Criar um caderno interativo com texto, código, gráficos e resultados. |
| Notebook | **Desktop** | Uma área de trabalho Linux no navegador, quando instalada | Abrir aplicativos gráficos dentro do ambiente do servidor. |
| Console | **Python 3** | Um console interativo Python | Testar comandos rapidamente, sem organizar tudo em um notebook. |
| Other | **Terminal** | Um terminal Linux | Executar comandos de sistema, instalar pacotes se permitido, listar arquivos, rodar scripts. |
| Other | **Text File** | Um arquivo `.txt` | Fazer anotações simples sem formatação especial. |
| Other | **Markdown File** | Um arquivo `.md` | Escrever texto formatado com títulos, listas, links, tabelas e fórmulas simples. |
| Other | **Python File** | Um arquivo `.py` | Criar um script Python tradicional. |
| Other | **Chat** | Painel de conversa com IA, quando configurado | Conversar com uma IA integrada ao JupyterLab, por exemplo via Jupyter AI ou provedor local. |
| Other | **Show Contextual Help** | Painel de ajuda contextual | Mostrar ajuda sobre funções, objetos ou comandos relacionados ao ponto em que o cursor está. |



## 3. Notebook — Python 3 (ipykernel)

O ícone **Python 3 (ipykernel)** cria um notebook Jupyter, isto é, um arquivo com extensão `.ipynb`.

Um notebook é composto por células:

- **células Markdown**, para explicações, títulos, fórmulas e textos;
- **células de código**, para executar Python;
- **saídas**, que podem conter números, tabelas, gráficos, mensagens ou erros.

### Quando usar um notebook?

Use notebook quando quiser:

- registrar uma experiência passo a passo;
- explicar raciocínios;
- combinar texto, código e resultados;
- criar material didático;
- documentar uma simulação;
- entregar uma atividade com execução e explicação juntas.

### Exemplo simples

Execute a célula abaixo para confirmar que o kernel Python está funcionando.


````python

import sys
from pathlib import Path

print("Python em uso:")
print(sys.version)
print()
print("Pasta atual:")
print(Path.cwd())
````


### Sinal importante: `[*]` ao lado da célula

Quando uma célula mostra `[*]`, isso significa que ela **está em execução**.

Enquanto a célula estiver com `[*]`, é melhor esperar.  
Se você executar outra célula que dependa do resultado da anterior, pode ocorrer erro ou resultado incompleto.

Quando a execução termina, o `[*]` é substituído por um número, como `[1]`, `[2]`, `[3]` etc.



## 4. Notebook — Desktop

Em alguns ambientes JupyterHub/JupyterLab, pode aparecer um ícone chamado **Desktop**.

No SimServ, quando esse recurso estiver instalado, ele costuma abrir uma área de trabalho Linux remota dentro do navegador.  
Isso permite usar programas gráficos que não cabem bem em uma célula de notebook, por exemplo:

- aplicativos com janelas;
- visualizadores gráficos;
- ferramentas educacionais;
- programas que usam interface de desktop.

### Atenção

O ícone **Desktop** não é parte obrigatória de toda instalação padrão do JupyterLab.  
Ele normalmente depende de extensões como `jupyter-remote-desktop-proxy`, de um servidor VNC e de um ambiente gráfico leve, como XFCE.

Se o botão **Desktop** não aparecer, não significa que o JupyterLab esteja quebrado.  
Significa apenas que esse recurso pode não estar instalado ou habilitado naquele ambiente.



## 5. Console — Python 3

O ícone **Python 3** na seção **Console** abre um console interativo.

Ele usa Python, mas não cria um notebook organizado com células Markdown e explicações.

### Quando usar o Console?

Use o console para:

- testar um comando rapidamente;
- verificar o valor de uma variável;
- experimentar uma função antes de colocá-la no notebook;
- executar pequenos trechos de código sem criar um arquivo `.ipynb`.

### Diferença entre Notebook e Console

| Recurso | Notebook | Console |
|---|---|---|
| Salva texto explicativo junto com código | Sim | Não, ou de forma limitada |
| Bom para relatório didático | Sim | Não |
| Bom para teste rápido | Sim, mas é mais organizado | Sim, é o uso principal |
| Produz arquivo `.ipynb` | Sim | Não necessariamente |
| Ideal para entregar atividade | Sim | Não |



## 6. Other — Terminal

O ícone **Terminal** abre um terminal Linux dentro do JupyterLab.

No terminal você pode usar comandos como:

```bash
pwd
ls
cd
python --version
pip list
```

### Para que serve?

O terminal é útil para:

- listar arquivos;
- criar pastas;
- executar scripts `.py`;
- verificar versões de programas;
- instalar pacotes, quando o ambiente permitir;
- diagnosticar problemas.

### Cuidado

O terminal é poderoso.  
Evite comandos destrutivos se não tiver certeza do que fazem.

Em especial, tenha cuidado com comandos como:

```bash
rm
rm -r
sudo
chmod
chown
```

No SimServ, muitas áreas compartilhadas são protegidas contra escrita.  
Normalmente o aluno deve trabalhar em sua pasta própria, como `/home/jovyan/work`.


````python

# Esta célula mostra uma forma segura de descobrir uma pasta de trabalho gravável.
# Ela NÃO altera arquivos importantes do sistema.

from pathlib import Path
import os

candidatas = [
    Path.home() / "work",
    Path.cwd(),
    Path.home()
]

for pasta in candidatas:
    if pasta.exists() and os.access(pasta, os.W_OK):
        pasta_trabalho = pasta
        break
else:
    pasta_trabalho = Path.cwd()

print("Pasta de trabalho escolhida para exemplos:")
print(pasta_trabalho)
print("Gravável?", os.access(pasta_trabalho, os.W_OK))
````


## 7. Other — Text File

O ícone **Text File** cria um arquivo de texto simples, geralmente com extensão `.txt`.

Use para:

- anotações simples;
- listas rápidas;
- rascunhos;
- pequenos registros;
- copiar e colar mensagens de erro.

Um arquivo `.txt` não tem células, kernel, gráficos embutidos nem execução de código.



## 8. Other — Markdown File

O ícone **Markdown File** cria um arquivo `.md`.

Markdown é uma linguagem simples para formatar texto.

Exemplos:

```markdown
# Título principal

## Subtítulo

- item 1
- item 2

**texto em negrito**

`código curto`

[link](https://jupyter.org)
```

Use arquivos Markdown para:

- roteiros de aula;
- documentação;
- README de projetos;
- instruções para colegas;
- descrição de uma simulação.

Diferença importante:

- Uma célula Markdown dentro de um notebook fica no arquivo `.ipynb`;
- Um arquivo Markdown independente fica em um arquivo `.md`.



## 9. Other — Python File

O ícone **Python File** cria um script Python tradicional, com extensão `.py`.

Use um arquivo `.py` quando quiser:

- criar funções reutilizáveis;
- organizar código longo;
- separar o código da explicação didática;
- executar o mesmo programa muitas vezes;
- importar funções em outros notebooks.

Exemplo:

```python
def saudacao(nome):
    return f"Olá, {nome}!"

print(saudacao("aluno"))
```

No terminal, um script pode ser executado com:

```bash
python nome_do_arquivo.py
```

Dentro do notebook, também é possível usar:

```python
%run nome_do_arquivo.py
```


````python

# Exemplo: criar arquivos de demonstração em uma pasta segura.
# Esta célula cria um .txt, um .md e um .py para mostrar a diferença entre eles.

from pathlib import Path
import os

base = Path.home() / "work"
if not (base.exists() and os.access(base, os.W_OK)):
    base = Path.cwd()

demo = base / "demo_launcher_jupyterlab"
demo.mkdir(exist_ok=True)

arquivo_txt = demo / "anotacao_simples.txt"
arquivo_md = demo / "roteiro_markdown.md"
arquivo_py = demo / "exemplo_script.py"

arquivo_txt.write_text("Este é um arquivo de texto simples.\\n", encoding="utf-8")

arquivo_md.write_text(
    "# Roteiro em Markdown\\n\\n"
    "- Este arquivo aceita títulos.\\n"
    "- Também aceita listas.\\n"
    "- Pode ser usado como documentação.\\n",
    encoding="utf-8"
)

arquivo_py.write_text(
    "def saudacao(nome):\\n"
    "    return f'Olá, {nome}! Este texto veio de um arquivo .py.'\\n\\n"
    "if __name__ == '__main__':\\n"
    "    print(saudacao('aluno'))\\n",
    encoding="utf-8"
)

print("Arquivos criados em:")
print(demo)
print()
for arq in [arquivo_txt, arquivo_md, arquivo_py]:
    print("-", arq.name)
````

````python

# Exemplo: executar o arquivo .py criado na célula anterior.
# O comando abaixo usa Python diretamente, sem depender de comandos perigosos de terminal.

import subprocess
import sys

resultado = subprocess.run(
    [sys.executable, str(arquivo_py)],
    capture_output=True,
    text=True
)

print("Saída do script:")
print(resultado.stdout)

if resultado.stderr:
    print("Mensagens de erro:")
    print(resultado.stderr)
````


## 10. Other — Chat

O ícone **Chat** aparece quando há uma extensão de conversa com IA instalada no JupyterLab.

Dependendo da instalação, esse recurso pode estar ligado a:

- Jupyter AI;
- um provedor externo;
- um LLM local;
- um servidor Ollama;
- outro serviço configurado pelo administrador.

### Para que serve?

Pode servir para:

- pedir explicação de erros;
- pedir revisão de código;
- gerar pseudocódigo;
- transformar narrativa em estrutura de simulação;
- pedir sugestões de melhoria;
- conversar com um modelo local do SimServ.

### Se o Chat aparecer, mas não responder

Isso pode ocorrer por vários motivos:

- nenhum provedor de IA foi configurado;
- o modelo local não está ativo;
- o endereço do serviço de IA está incorreto;
- o aluno não tem permissão para acessar o serviço;
- a extensão está instalada, mas sem configuração completa.

No SimServ, o ideal é ter uma célula de diagnóstico simples para testar o serviço local de IA separadamente.



## 11. Other — Show Contextual Help

O ícone **Show Contextual Help** abre um painel de ajuda contextual.

A função esperada é mostrar informações sobre o objeto, função ou comando relacionado ao ponto em que o cursor está.

Por exemplo, se você estiver com o cursor em uma função Python como:

```python
print
len
sum
```

ou em uma função importada como:

```python
math.sin
pandas.DataFrame
```

o painel pode mostrar documentação ou assinatura da função.

### Por que o painel pode estar vazio?

O painel pode aparecer vazio quando:

1. não há uma célula ou editor ativo;
2. o cursor não está sobre uma função, variável ou objeto reconhecível;
3. o kernel ainda não foi iniciado;
4. a célula ainda não foi executada;
5. o pacote ainda não foi importado;
6. a extensão de ajuda/inspeção está desabilitada;
7. o ambiente não tem um servidor de linguagem instalado para ajuda mais rica.

### Teste simples

1. Abra o **Show Contextual Help**.
2. Execute a próxima célula.
3. Clique dentro da célula, sobre `math.sin`.
4. Use o menu de ajuda contextual, se necessário.
5. Veja se o painel passa a mostrar informação.


````python

import math

# Depois de abrir o painel "Show Contextual Help",
# clique com o cursor sobre "math.sin" nesta linha:
math.sin(0.5)
````


## 12. Como verificar se a ajuda contextual e o LSP estão instalados

A ajuda contextual básica pode funcionar com o kernel Python/IPython.  
Para recursos mais ricos de edição, autocompletar, navegação e inspeção, muitas instalações usam `jupyterlab-lsp` junto com um servidor de linguagem Python, como `python-lsp-server`.

A célula abaixo apenas verifica se alguns pacotes parecem estar instalados no ambiente atual.


````python

import importlib.util

pacotes = {
    "jupyterlab_lsp": "jupyterlab-lsp",
    "jupyter_lsp": "jupyter-lsp",
    "pylsp": "python-lsp-server",
    "jedi": "jedi",
}

for modulo, pacote in pacotes.items():
    encontrado = importlib.util.find_spec(modulo) is not None
    print(f"{pacote:25s} -> {'encontrado' if encontrado else 'não encontrado'}")
````


### Comandos úteis de diagnóstico

No **Terminal** do JupyterLab, o administrador ou usuário autorizado pode testar:

```bash
jupyter labextension list
jupyter server extension list
python -m pip show jupyterlab-lsp
python -m pip show jupyter-lsp
python -m pip show python-lsp-server
```

Em um ambiente Docker/JupyterHub como o SimServ, o mais adequado é instalar esses componentes na **imagem do usuário**, e não manualmente dentro da sessão de cada aluno.

Exemplo para um `Dockerfile.user`:

```dockerfile
RUN pip install --no-cache-dir \
    jupyterlab-lsp \
    "python-lsp-server[all]"
```

Depois disso, normalmente é necessário reconstruir a imagem e reiniciar os servidores dos usuários.

Exemplo conceitual:

```bash
docker compose build
docker compose up -d
```

> Observação: os nomes dos serviços Docker podem variar no SimServ. Use os nomes reais do seu `docker-compose.yml`.
