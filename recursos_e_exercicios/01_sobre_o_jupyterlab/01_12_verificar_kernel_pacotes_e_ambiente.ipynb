## Objetivo

Verificar informações básicas do ambiente Python usado pelo notebook.

```python
import sys
import platform
from pathlib import Path

print("Python:", sys.version)
print("Executável Python:", sys.executable)
print("Sistema:", platform.platform())
print("Pasta atual:", Path.cwd())
print("Pasta home:", Path.home())
```

```python
bibliotecas = [
    "numpy",
    "pandas",
    "matplotlib",
    "requests",
    "IPython",
    "ipykernel",
]

print("Verificação de bibliotecas:\n")
for nome in bibliotecas:
    try:
        modulo = __import__(nome)
        versao = getattr(modulo, "__version__", "versão não informada")
        print(f"{nome:12} OK      {versao}")
    except Exception as erro:
        print(f"{nome:12} ERRO    {erro}")
```

```python
import os
from pathlib import Path

for pasta in [Path.home(), Path.home() / "work", Path.home() / "laboratorio", Path.cwd()]:
    if pasta.exists():
        print(f"{str(pasta):45} escrita={os.access(pasta, os.W_OK)}")
```
