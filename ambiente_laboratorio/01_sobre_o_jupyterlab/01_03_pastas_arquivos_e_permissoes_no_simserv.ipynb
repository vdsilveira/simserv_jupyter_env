## Objetivo

Entender a diferença entre pastas de leitura e pastas de trabalho.

Em geral, no SimServ haverá pastas com materiais prontos e pastas onde o aluno pode salvar seus próprios arquivos.

## Modelo conceitual das pastas

| Pasta possível | Papel didático | Escrita pelo aluno? |
|---|---|---|
| `/sobre_ambientes_simserv` | Orientação sobre o ambiente | Não |
| `/home/jovyan/shared` | Materiais compartilhados pelo professor | Normalmente não |
| `/home/jovyan/exercicios_e_recursos` | Nome didático para materiais compartilhados | Normalmente não |
| `/home/jovyan/work` | Área de trabalho do aluno | Sim |
| `/home/jovyan/laboratorio` | Nome didático para área de trabalho | Sim |

```python
from pathlib import Path
import os

pastas = [
    Path.cwd(),
    Path.home(),
    Path.home() / "work",
    Path.home() / "laboratorio",
    Path.home() / "shared",
    Path.home() / "exercicios_e_recursos",
    Path("/sobre_ambientes_simserv"),
]

print("Verificação de pastas e permissões\n")
for pasta in pastas:
    existe = pasta.exists()
    leitura = os.access(pasta, os.R_OK) if existe else False
    escrita = os.access(pasta, os.W_OK) if existe else False
    print(f"{str(pasta):45} existe={existe!s:5} leitura={leitura!s:5} escrita={escrita!s:5}")
```

## Interpretação

- Se `escrita=False`, não tente salvar seus trabalhos nessa pasta.
- Se o notebook está em uma pasta de somente leitura, use **Save Notebook As...** para salvar uma cópia na área de trabalho.
- Se a área de trabalho também aparecer como não gravável, envie o diagnóstico ao professor ou administrador.
