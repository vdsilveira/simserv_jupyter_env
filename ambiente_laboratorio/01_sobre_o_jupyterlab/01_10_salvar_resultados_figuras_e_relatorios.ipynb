## Objetivo

Criar pastas de resultados, salvar uma figura e salvar um relatório em texto.

```python
from pathlib import Path
import os

def encontrar_area_de_trabalho():
    """Procura uma pasta gravável para salvar resultados do aluno."""
    candidatas = [
        Path.cwd().parent.parent / "work",
        Path.home() / "laboratorio",
        Path.home() / "work",
        Path("/home/jovyan/laboratorio"),
        Path("/home/jovyan/work"),
        Path.cwd(),
    ]
    for pasta in candidatas:
        try:
            if pasta.exists() and os.access(pasta, os.W_OK):
                return pasta
        except Exception:
            pass
    raise RuntimeError(
        "Não encontrei uma pasta gravável. Abra uma cópia do notebook na sua área de trabalho."
    )

AREA_TRABALHO = encontrar_area_de_trabalho()
print("Área de trabalho escolhida:", AREA_TRABALHO)
```

```python
from pathlib import Path

base = AREA_TRABALHO / "resultados_exemplo_simserv"
figuras = base / "figuras"
relatorios = base / "relatorios"

figuras.mkdir(parents=True, exist_ok=True)
relatorios.mkdir(parents=True, exist_ok=True)

print("Pasta principal:", base)
print("Pasta de figuras:", figuras)
print("Pasta de relatórios:", relatorios)
```

```python
import matplotlib.pyplot as plt

x = [0, 1, 2, 3]
y = [0, 1, 4, 9]

plt.plot(x, y, marker="o")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Figura salva pelo notebook")
plt.grid(True)

caminho_figura = figuras / "grafico_exemplo.png"
plt.savefig(caminho_figura, dpi=150, bbox_inches="tight")
plt.show()

print("Figura salva em:", caminho_figura)
```

```python
texto = """
Relatório simples
=================

Este arquivo foi criado a partir de um notebook no SimServ.
Ele pode ser usado como exemplo para salvar observações, resultados e conclusões.
""".strip()

caminho_relatorio = relatorios / "relatorio_exemplo.txt"
caminho_relatorio.write_text(texto, encoding="utf-8")

print("Relatório salvo em:", caminho_relatorio)
```
