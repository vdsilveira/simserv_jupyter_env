## Regras simples para evitar problemas

| Recomendação | Por quê? |
|---|---|
| Use nomes sem acentos e sem espaços | Evita problemas de compatibilidade em scripts e caminhos. |
| Salve versões `V1`, `V2`, `V3` | Permite voltar a uma versão anterior. |
| Execute as células em ordem | Evita variáveis indefinidas e resultados confusos. |
| Não edite material em pasta somente leitura | Preserve o gabarito e evite erro de permissão. |
| Leia a última linha da mensagem de erro | Muitas vezes ela resume o problema. |
| Reinicie o kernel quando o estado ficar confuso | Limpa variáveis antigas e ajuda a reproduzir resultados. |
