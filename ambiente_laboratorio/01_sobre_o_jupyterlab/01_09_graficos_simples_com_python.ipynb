## Objetivo

Criar gráficos simples usando `matplotlib`.

```python
import matplotlib.pyplot as plt

tempos = [0, 1, 2, 3, 4]
posicoes = [0, 1, 4, 9, 16]

plt.plot(tempos, posicoes, marker="o")
plt.xlabel("Tempo")
plt.ylabel("Posição")
plt.title("Exemplo de gráfico simples")
plt.grid(True)
plt.show()
```

## Gráfico de barras

```python
categorias = ["A", "B", "C"]
valores = [10, 15, 7]

plt.bar(categorias, valores)
plt.xlabel("Categoria")
plt.ylabel("Valor")
plt.title("Exemplo de gráfico de barras")
plt.show()
```

## Histograma simples

```python
medidas = [1.0, 1.1, 1.0, 1.3, 1.2, 1.1, 1.4, 1.2, 1.3, 1.2]

plt.hist(medidas, bins=4)
plt.xlabel("Medida")
plt.ylabel("Frequência")
plt.title("Histograma simples")
plt.show()
```
