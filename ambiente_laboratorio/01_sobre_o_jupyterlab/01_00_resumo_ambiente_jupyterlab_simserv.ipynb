## Para que serve este conjunto de notebooks?

Esta série apresenta o ambiente JupyterLab usado no SimServ. O objetivo é ajudar o aluno a entender:

- o que é um notebook;
- como executar células;
- onde estão os materiais do professor;
- onde o aluno deve salvar seus próprios arquivos;
- como verificar se o ambiente está funcionando;
- como pedir ajuda com informações úteis quando houver problema.

## Vocabulário mínimo

| Termo | Significado prático |
|---|---|
| JupyterLab | Ambiente gráfico no navegador para trabalhar com notebooks, arquivos, terminais e dados. |
| Notebook | Arquivo com células de texto, código e resultados. Normalmente termina em `.ipynb`. |
| Célula Markdown | Célula usada para explicações, títulos, listas, tabelas e fórmulas. |
| Célula Code | Célula usada para executar comandos Python. |
| Kernel | Processo que executa o código Python do notebook. |
| Pasta somente leitura | Pasta onde você pode abrir e executar, mas não deve editar diretamente. |
| Área de trabalho | Pasta onde você pode salvar seus arquivos, resultados e versões. |

```python
import sys
import platform
from pathlib import Path

print("Teste inicial do ambiente SimServ")
print("Python:", sys.version.split()[0])
print("Sistema:", platform.platform())
print("Pasta atual:", Path.cwd())
print("Pasta home:", Path.home())
```

## Mensagem importante
