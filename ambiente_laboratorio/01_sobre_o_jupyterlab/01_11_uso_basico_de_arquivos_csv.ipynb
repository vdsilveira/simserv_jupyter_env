## Objetivo

Salvar e ler dados em formato CSV. Esse formato é muito usado para resultados de simulações e experimentos.

````python
from pathlib import Path
import os

def encontrar_area_de_trabalho():
    """Procura uma pasta gravável para salvar resultados do aluno."""
    candidatas = [
        Path.home() / "laboratorio",
        Path.home() / "work",
        Path("/home/jovyan/laboratorio"),
        Path("/home/jovyan/work"),
        Path.cwd(),
    ]
    for pasta in candidatas:
        try:
            if pasta.exists() and os.access(pasta, os.W_OK):
                return pasta
        except Exception:
            pass
    raise RuntimeError(
        "Não encontrei uma pasta gravável. Abra uma cópia do notebook na sua área de trabalho."
    )

AREA_TRABALHO = encontrar_area_de_trabalho()
print("Área de trabalho escolhida:", AREA_TRABALHO)
````

````python
try:
    import pandas as pd
except Exception as erro:
    raise RuntimeError("Este notebook precisa da biblioteca pandas para a parte de CSV.") from erro

dados = pd.DataFrame({
    "tempo": [0, 1, 2, 3],
    "valor": [10, 12, 15, 18],
})

dados
````

````python
saida = AREA_TRABALHO / "dados_exemplo_simserv.csv"
dados.to_csv(saida, index=False)

print("Arquivo CSV salvo em:")
print(saida)
````

````python
dados_lidos = pd.read_csv(saida)
dados_lidos
````
