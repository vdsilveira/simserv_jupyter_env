## Objetivo

Conhecer alguns erros comuns em Python sem deixar o notebook travado.

As células abaixo usam `try/except` para capturar o erro e mostrar uma explicação.

```python
try:
    print(nome_que_nao_existe)
except Exception as erro:
    print("Tipo do erro:", type(erro).__name__)
    print("Mensagem:", erro)
    print("Explicação: a variável foi usada antes de receber um valor.")
```

```python
try:
    resultado = 10 / 0
except Exception as erro:
    print("Tipo do erro:", type(erro).__name__)
    print("Mensagem:", erro)
    print("Explicação: não é possível dividir por zero.")
```

```python
try:
    lista = [1, 2, 3]
    print(lista[10])
except Exception as erro:
    print("Tipo do erro:", type(erro).__name__)
    print("Mensagem:", erro)
    print("Explicação: a lista tem apenas três elementos, com índices 0, 1 e 2.")
```

```python
try:
    import biblioteca_que_nao_existe
except Exception as erro:
    print("Tipo do erro:", type(erro).__name__)
    print("Mensagem:", erro)
    print("Explicação: o Python não encontrou essa biblioteca instalada no ambiente atual.")
```

## Como ler uma mensagem de erro

Procure três informações:

1. **nome do erro**, por exemplo `NameError`;
2. **linha indicada**, onde o problema apareceu;
3. **mensagem final**, que costuma resumir a causa.
