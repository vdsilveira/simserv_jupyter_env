## Objetivo

Representar pequenas sequências de dados e organizá-las em tabela.

```python
tempos = [0, 1, 2, 3, 4]
posicoes = [0, 1, 4, 9, 16]

for t, x in zip(tempos, posicoes):
    print("tempo =", t, "posição =", x)
```

## Tabela com pandas

A biblioteca `pandas` é muito usada para trabalhar com tabelas.

```python
try:
    import pandas as pd

    dados = pd.DataFrame({
        "tempo": tempos,
        "posicao": posicoes,
    })
    display(dados)
except Exception as erro:
    print("Não foi possível usar pandas neste ambiente.")
    print("Erro:", erro)
```

```python
# Cálculo simples de diferenças entre posições consecutivas
deslocamentos = []

for i in range(1, len(posicoes)):
    deslocamentos.append(posicoes[i] - posicoes[i-1])

print("Deslocamentos:", deslocamentos)
```
