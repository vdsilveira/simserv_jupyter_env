## 4. Gerar resposta com IA

Execute a célula abaixo para que a IA analise o contexto e gere uma resposta automaticamente.


```python
from dotenv import load_dotenv
load_dotenv()
import os, json
from urllib.request import Request, urlopen

MODELO = os.getenv('OLLAMA_MODEL', 'ministral-3:3b')
BASE_URL = os.getenv('OLLAMA_BASE_URL', 'https://api.ollama.com').rstrip('/')
API_KEY = os.getenv('OLLAMA_API_KEY')

prompt = '''Voce e um assistente de um curso de sistemas de muitos agentes.
Responda em portugues, de forma clara e objetiva (maximo 5 paragrafos).

Pergunta: Com base nos requisitos de hardware e software discutidos no Topico 06.03,
descreva os componentes essenciais de uma infraestrutura computacional para o estudo de
sistemas complexos. Em particular, explique:

1. Quais as funcoes do JupyterLab como instrumento de aprendizado e ferramenta de desenvolvimento?
2. Como os grandes modelos de linguagem (LLM) se integram a esse ambiente?
3. Qual a importancia da abordagem multicamadas (SO -> ambiente -> bibliotecas -> LLM)?

Contexto:
- Hardware necessario: CPU, RAM, armazenamento, GPU (opcional para LLMs locais).
- Software: SO (Linux/Windows/macOS), JupyterLab como ambiente integrado.
- Bibliotecas: numpy, scipy, networkx, matplotlib, pandas, simpy, mesa (ABM).
- LLM: Ollama Cloud (API remota) ou modelos locais para geracao de codigo/analise.
- JupyterLab permite celulas de codigo, markdown, visualizacoes, tudo integrado.'''

headers = {
    'Content-Type': 'application/json',
    'User-Agent': 'Mozilla/5.0',
    'Accept': 'application/json',
}
if API_KEY:
    headers['Authorization'] = f'Bearer {API_KEY}'

payload = {
    'model': MODELO,
    'messages': [{'role': 'user', 'content': prompt}],
    'stream': False,
    'options': {'temperature': 0.3, 'top_p': 0.9, 'num_predict': 1000},
}

try:
    req = Request(BASE_URL + '/api/chat', data=json.dumps(payload).encode('utf-8'), headers=headers, method='POST')
    with urlopen(req, timeout=60) as resp:
        dados = json.loads(resp.read().decode('utf-8'))
    resposta = dados.get('message', {}).get('content', '').strip()
    if resposta:
        resposta_pergunta_3 = resposta
        print('Resposta gerada com sucesso!')
    else:
        print('IA retornou resposta vazia.')
except Exception as e:
    print(f'Erro ao consultar IA: {type(e).__name__}: {e}')

print()
print('Resposta atual:')
print(resposta_pergunta_3)
```

## 5. Exportar para o Moodle

Esta célula também mescla os PDFs dos 3 exercícios em um único PDF combinado.


```python
import os
import shutil
import subprocess
from dotenv import load_dotenv
load_dotenv()

NOME = os.getenv('NOME', 'nao informado')
RA = os.getenv('RA', 'nao informado')
MODELO = os.getenv('OLLAMA_MODEL', 'nao informado')

from datetime import datetime

agora = datetime.now().strftime('%Y-%m-%d %H:%M')

PATH_MD = os.path.join(PASTA_AULA, 'aula_06_exercicio_03.md')
with open(PATH_MD, 'w', encoding='utf-8') as f:
    f.write('# Aula 06 — Exercicio 03: Infraestrutura computacional\n\n')
    f.write(f'_Executado em: {agora}_\n\n')
    f.write('---\n\n')
    f.write(f'**Aluno:** {NOME} (RA: {RA})\\\n\\n')
    f.write(f'**Modelo LLM:** {MODELO}\\\n\\n')
    f.write('**Processo:** Os 3 notebooks geram PDFs individuais que sao mesclados\\\n')
    f.write('em um unico PDF via pdfunite. O aluno revisa e pode alterar\\\n')
    f.write('manualmente a resposta da IA antes da exportacao.\\\n\\n')
    f.write('---\\\n\\n')
    f.write('## Resposta do aluno\n\n')
    f.write(resposta_pergunta_3.strip())
    f.write('\n')

PATH_PDF = PATH_MD.replace('.md', '.pdf')
if shutil.which('pandoc'):
    subprocess.run(['pandoc', PATH_MD, '-o', PATH_PDF, '--pdf-engine=xelatex', '--resource-path', PASTA_AULA, '-V', 'mainfont=DejaVu Serif'], check=True)
else:
    print('Aviso: pandoc nao instalado. Execute ./setup.sh para instalar.')
print(f'Markdown: {os.path.abspath(PATH_MD)}')
print(f'PDF:      {os.path.abspath(PATH_PDF)}')

# Mesclar PDFs individuais em um unico PDF
if shutil.which('pdfunite'):
    import glob
    pdfs = sorted(glob.glob(os.path.join(PASTA_AULA, 'aula_06_exercicio_??.pdf')))
    if len(pdfs) == 3:
        PASTA_RESPOSTAS = os.path.join(PASTA_MOODLE, 'respostas')
        os.makedirs(PASTA_RESPOSTAS, exist_ok=True)
        PATH_COMBINADO = os.path.join(PASTA_RESPOSTAS, 'aula_06_completo.pdf')
        subprocess.run(['pdfunite'] + pdfs + [PATH_COMBINADO], check=True)
        print(f'PDF combinado: {os.path.abspath(PATH_COMBINADO)}')
    else:
        print(f'Aviso: encontrados {len(pdfs)} PDFs, esperados 3. Execute todos os exercicios primeiro.')
```
